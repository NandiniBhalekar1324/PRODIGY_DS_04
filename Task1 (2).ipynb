Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv(r"C:\\Users\\Hp\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_9949\\API_SP.POP.TOTL_DS2_en_csv_v2_9949.csv", skiprows=4)
```

```python
# Display the first few rows 
print(df.head())
# Display basic information
print(df.info())
```

```python
# Filter data for the year 2020
df_2020 = df[['Country Name', '2020']].dropna()  # Remove rows with no data for 2020
```

```python

# Create a bar chart
plt.figure(figsize=(40, 8))
plt.bar(df_2020['Country Name'], df_2020['2020'], color='red')
plt.xlabel("Countries")
plt.ylabel("Population in year 2020")
plt.title("Population by Country in year 2020")
plt.xticks(rotation=70, ha='left') 
plt.tight_layout()
plt.show()
```
