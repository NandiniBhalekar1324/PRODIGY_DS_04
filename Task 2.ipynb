Perform data cleaning and exploratory data analysis (EDA) on a dataset of your choice, such as the Titanic dataset from Kaggle. Explore the relationships between variables and identify patterns and trends in the data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming you have downloaded the Titanic dataset as a CSV file
df = pd.read_csv("C:\\Users\\Hp\\Downloads\\train (1).csv")
```

```python
print(df.isnull().sum())
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
# Box plot for outlier detection in 'Age'
sns.boxplot(x=df['Age'])
plt.show()
```

```python
print(df.describe())
```

```python
# Age Distribution Histogram 
plt.hist(df['Age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

```python
#Gender Distribution
sns.countplot(x='Sex', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
# Age Distribution by Passenger Class
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
plt.show()
```

```python
print(df.dtypes)
```

```python
# Correlation Matrix
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

Insights from EDA
Survival Rate: The overall survival rate is approximately 38%.
Impact of Passenger Class: Passengers in higher classes (1st and 2nd) had significantly higher survival rates compared to those in 3rd class.
Gender Impact: Women had a higher survival rate compared to men.
Age Impact: Children and younger adults had a higher chance of survival.
Family Size: Passengers traveling alone had a lower survival rate compared to those traveling with family.
Conclusion
By performing data cleaning and EDA, we gained valuable insights into the factors influencing survival on the Titanic. These insights can be used to build predictive models to predict survival for future scenarios.
